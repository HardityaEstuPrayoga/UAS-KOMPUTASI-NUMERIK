```python
import numpy as np
import matplotlib.pyplot as plt

# Mendefinisikan vektor
v_P = np.array([5, 10, -6])
v_A = np.array([10, -5, 4])

# Menghitung v_P + v_A
v_P_plus_v_A = v_P + v_A

# Menghitung v_P - v_A
v_P_minus_v_A = v_P - v_A

# Menampilkan hasil
print("v_P + v_A = ", v_P_plus_v_A[0], "i +", v_P_plus_v_A[1], "j +", v_P_plus_v_A[2], "k")
print("v_P - v_A = ", v_P_minus_v_A[0], "i +", v_P_minus_v_A[1], "j +", v_P_minus_v_A[2], "k")

# Plotting vectors
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Vektor awal
ax.quiver(0, 0, 0, v_P[0], v_P[1], v_P[2], color='r', label='v_P (5i + 10j - 6k)')
ax.quiver(0, 0, 0, v_A[0], v_A[1], v_A[2], color='g', label='v_A (10i - 5j + 4k)')

# Hasil penjumlahan dan pengurangan
ax.quiver(0, 0, 0, v_P_plus_v_A[0], v_P_plus_v_A[1], v_P_plus_v_A[2], color='b', label='v_P + v_A (15i + 5j - 2k)')
ax.quiver(0, 0, 0, v_P_minus_v_A[0], v_P_minus_v_A[1], v_P_minus_v_A[2], color='y', label='v_P - v_A (-5i + 15j - 10k)')

# Setting plot limits
ax.set_xlim([-10, 20])
ax.set_ylim([-10, 20])
ax.set_zlim([-10, 10])

# Labeling the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Adding a legend
ax.legend()

# Showing the plot
plt.show()
```
